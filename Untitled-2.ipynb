## Statistical analysis

Q1) Connection between family history and cancer stage

Type of Variables:

family_history: Categorical (Yes/No)

cancer_stage: Categorical (Stage I, II, III, IV)

Hypotheses:

H₀ (null): There is no association between family history and cancer stage.

H₁ (alt): There is a significant association between family history and cancer stage.

Test: Chi-square test

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create the contingency table
contingency = pd.crosstab(df['family_history'], df['cancer_stage'])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency)

# Print results
print("Chi-square Test Results")
print(f"Chi-square statistic: {chi2}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p}")

# Interpret the result
alpha = 0.05
if p < alpha:
    print("There is a statistically significant association between family history and cancer stage (Reject H0).")
else:
    print("There is no statistically significant association between family history and cancer stage (Fail to reject H0).")
```

 Q7) Do BMI, cholesterol level, hypertension, asthma, and cirrhosis significantly predict survival outcomes in lung cancer patients?



Dependent Variable: survived

Predictors: BMI, cholesterol_level, hypertension, asthma, cirrhosis

H₀: BMI, cholesterol level, hypertension, asthma, and cirrhosis do not significantly predict survival in lung cancer patients.

(H₁): At least one of these variables significantly predicts survival in lung cancer patients.

Test: Logistic Regression

```python
import pandas as pd
import statsmodels.api as sm

# Define predictors and outcome variable
X = df[['bmi', 'cholesterol_level', 'hypertension', 'asthma', 'cirrhosis']]
y = df['survived']  # 0 = died, 1 = survived

# Add constant (intercept) to the model
X = sm.add_constant(X)

# Fit logistic regression model
model = sm.Logit(y, X).fit()

# Print model summary
print(model.summary())

# Interpret model significance
alpha = 0.05
if model.llr_pvalue < alpha:
    print("The model is statistically significant (Reject H₀): At least one variable predicts survival.")
else:
    print("The model is not statistically significant (Fail to reject H₀): These variables do not meaningfully predict survival.")
```

The logistic regression analysis was conducted to determine whether BMI, cholesterol level, hypertension, asthma, and cirrhosis predict survival in lung cancer patients. The model, based on 890,000 observations, was not statistically significant overall (LLR p-value = 0.2424), indicating that these variables do not collectively explain a meaningful variation in survival outcomes. Additionally, individual predictors — including BMI (p = 0.876), cholesterol (p = 0.829), and hypertension (p = 0.950) — showed no significant effect, while asthma (p = 0.071) and cirrhosis (p = 0.054) were marginally non-significant. The model's pseudo R-squared value was extremely low (0.00000715), suggesting poor explanatory power. Therefore, we conclude that these health factors do not significantly influence survival in this dataset.


These results indicate that BMI, cholesterol level, hypertension, asthma, and cirrhosis alone do not significantly predict survival outcomes in lung cancer patients within this dataset. While cirrhosis and asthma showed very small, borderline effects, the overall impact of these factors was minimal. This suggests that other variables—not included in this analysis—likely play a larger role in influencing lung cancer survival. Therefore, it’s important to consider additional clinical, demographic, or treatment-related factors when studying predictors of lung cancer outcomes.

# Survival Analysis

```python
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

kmf = KaplanMeierFitter()

T = df['treatment_duration_years']  # duration in years
E = 1 - df['survived']               # event = death (assuming survived=1 means alive)

kmf.fit(T, event_observed=E)

kmf.plot_survival_function()
plt.title('Kaplan-Meier Survival Curve (Years)')
plt.xlabel('Years since Diagnosis')
plt.ylabel('Survival Probability')
plt.show()
```

The Kaplan-Meier survival curve shows the probability of survival over a 2-year period following a lung cancer diagnosis. Initially, the survival probability is high (close to 1.0), indicating that most patients survive the first year. However, there is a sharp drop around the **1-year mark**, reducing the survival probability to approximately **0.47**, suggesting that a significant portion of patients die after one year. The curve remains flat until close to the **2-year mark**, where it drops steeply again to around **0.05**, indicating that very few patients survive beyond two years. Overall, the curve reflects a steep decline in survival after the first year of diagnosis.


```python
plt.figure(figsize=(10,6))
for stage in df['cancer_stage'].unique():
    ix = df['cancer_stage'] == stage
    kmf.fit(T[ix], event_observed=E[ix], label=str(stage))
    kmf.plot_survival_function()

plt.title('Survival Curve by Cancer Stage (Years)')
plt.xlabel('Years since Diagnosis')
plt.ylabel('Survival Probability')
plt.show()
```

This Kaplan-Meier survival curve shows the **survival probability over 2 years** for lung cancer patients, segmented by **cancer stage at diagnosis (Stage I to IV)**.

### **Interpretation**:

1. **Initial Survival (Year 0)**:

   * All stages start with high survival probabilities (close to 1), indicating that most patients are alive at diagnosis.
   * **Stage I** starts the highest, followed by **Stage II**, **Stage III**, and **Stage IV** — as expected.

2. **Sharp Drop at Year 1**:

   * All stages experience a significant drop in survival at the 1-year mark.
   * The magnitude of the drop is **steeper for higher stages**, especially **Stage IV**, indicating more aggressive disease.
